# 4.10.3 The Instacart officers are interested in comparing customer behavior in different geographic areas. Create a regional segmentation of the data. You’ll need to create a “Region” column based on the “State” column from your customers data set. Determine whether there’s a difference in spending habits between the different U.S. regions. 

```python
#  checking state names and value counts

df['state'].value_counts(dropna = False)
```

```python
#  grouping states into regions and applying the desired labels

result = []

for value in df['state']:
    if (value == 'Maine') or (value == 'New Hampshire') or (value == 'Vermont') or (value == 'Massachusetts') or (value == 'Rhode Island') or (value == 'Connecticut') or (value == 'New York') or (value == 'Pennsylvania') or (value =='New Jersey'):
        result.append('Northeast')
    elif (value == 'Wisconsin') or (value == 'Michigan') or (value == 'Illinois') or (value == 'Indiana') or (value == 'Ohio') or (value == 'North Dakota') or (value == 'South Dakota') or (value == 'Nebraska') or (value == 'Kansas') or (value == 'Minnesota') or (value == 'Iowa') or (value == 'Missouri'):
        result.append('Midwest')
    elif (value == 'Deleware') or (value == 'Maryland') or (value == 'District of Columbia') or (value == 'Virginia') or (value == 'West Virginia') or (value == 'North Carolina') or (value == 'South Carolina') or (value == 'Georgia') or (value == 'Florida') or (value == 'Kentucky') or (value == 'Tennessee') or (value == 'Mississippi') or (value == 'Alabama') or (value == 'Oklahoma') or (value == 'Texas') or (value == 'Arkansas') or (value == 'Louisiana'):
        result.append('South')
    else:
        result.append('West')
```

```python
#  defining income groups (arbitrary amounts)

#  0-25k
#  25-40k
#  40-55k
#  55-75k
#  75-100k
#  100-150k
#  150k-250k
#  250k-500k
#  500k+
```

```python
#  creating income groups

income = []

for value in df['income']:
    if value <= 25000: 
        income.append('0-25k')
    elif (value > 25000) and (value <= 40000):
        income.append('025-40k')
    elif (value > 40000) and (value <= 55000):
        income.append('040-55k')
    elif (value > 55000) and (value <= 75000):
        income.append('055-75k')
    elif (value > 75000) and (value <= 100000):
        income.append('075-100k')
    elif (value > 100000) and (value <= 150000):
        income.append('100-150k')
    elif (value > 150000) and (value <= 250000):
        income.append('150k-250k')
    elif (value > 250000) and (value <= 500000):
        income.append('250k-500k')
    elif (value > 500000):
        income.append('500k+')
    else: 
        income.append('other')
```

```python
#  creating the new income_group column

df['income_group'] = income
```

```python
#  counting entries of new income_group column

df['income_group'].value_counts(dropna = False).sort_index()
```
