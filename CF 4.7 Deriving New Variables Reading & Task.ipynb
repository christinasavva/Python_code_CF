```python
#  define a function (ie. price_label)

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

# Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
#  as per the value counts above, Sundays (1) are the second busiest and Tuesday (3) the second least busy days. I have therefore specified them in the code. I also amended the name to 'Busiest days'.

result = []

for value in df_ords_prods_merged["order_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days")
  elif value == 4:
    result.append("Least busy")
  elif value == 3:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_days'] = result
```

# When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#  finding out the busiest order_hour_of_day

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#  grouping the order_hour_of_day into the buckets and applying the desired labels (Most orders = min 7.6% of total orders, Avg Orders = 1.2% to 6.4% of total orders, Fewest Orders = under 0.9% of total orders)

result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 10:
    result.append("Most orders")
  elif value == 11:
    result.append("Most orders")
  elif value == 14:
    result.append("Most orders")
  elif value == 15:
    result.append("Most orders")
  elif value == 13:
    result.append("Most orders")
  elif value == 12:
    result.append("Most orders")
  elif value == 16:
    result.append("Most orders")
  elif value == 9:
    result.append("Most orders")
  elif value == 17:
    result.append("Average orders")
  elif value == 8:
    result.append("Average orders")
  elif value == 18:
    result.append("Average orders")
  elif value == 19:
    result.append("Average orders")
  elif value == 20:
    result.append("Average orders")
  elif value == 7:
    result.append("Average orders")
  elif value == 21:
    result.append("Average orders")
  elif value == 22:
    result.append("Average orders")
  elif value == 23:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

```python
df_ords_prods_merged['busiest_period_of_day'] = result
```
